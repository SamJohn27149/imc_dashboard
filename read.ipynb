```python
import json

# 按 timestamp 遍历：解析 lambdaLog → 字典；再按品种、bid 档位键展开
# 每条记录：timestamp、product、bid 侧 JSON 里的键（报价档位）与对应 volume
lambda_log_bid_rows = []
# 每个 timestamp 仅一行：对表按时间排序后单次 iterrows，避免每个 ts 再过滤子表
for _, row in logs_df.sort_values("timestamp").iterrows():
    timestamp = row["timestamp"]
    raw = row["lambdaLog"]
    if raw is None or raw == "" or (isinstance(raw, float) and pd.isna(raw)):
        continue
    if isinstance(raw, dict):
        d = raw
    else:
        try:
            d = json.loads(raw)
        except (json.JSONDecodeError, TypeError):
            continue
    if not isinstance(d, dict):
        continue
    for product, value in d.items():
        if not isinstance(value, dict):
            continue
        bid_map = value.get("bid") or {}
        if not isinstance(bid_map, dict):
            continue
        for bid_key, bid_vol in bid_map.items():
            lambda_log_bid_rows.append(
                {
                    "timestamp": timestamp,
                    "product": product,
                    "bid_quote_key": bid_key,
                    "bid_volume": bid_vol,
                }
            )

lambda_log_bid_df = pd.DataFrame(lambda_log_bid_rows)

# 若需要「每条是一组三个小字典」的列表形式（便于和原描述对齐），可用下面这种结构：
# bundle = [
#     {"timestamp": timestamp},
#     {"product": product},
#     {"bid_quote_key": bid_key, "bid_volume": bid_vol},
# ]
lambda_log_bid_bundles = [
    [
        {"timestamp": r["timestamp"]},
        {"product": r["product"]},
        {"bid_quote_key": r["bid_quote_key"], "bid_volume": r["bid_volume"]},
    ]
    for r in lambda_log_bid_rows
]

lambda_log_bid_df.head(10)
```

```python
SPECIAL_PRODUCT = ["A","B"]

def _expand_logs_for_special_products(logs_df: pd.DataFrame) -> pd.DataFrame:
    if logs_df.empty:
        return logs_df
    original = logs_df.copy()
    n = len(original)
    if  SPECIAL_PRODUCT:
        k = len(SPECIAL_PRODUCT)
        expanded = original.loc[original.index.repeat(k)].reset_index(drop=True)
        expanded["product"] = SPECIAL_PRODUCT * n
    return expanded

expanded_logs_df = _expand_logs_for_special_products(logs_df)
expanded_logs_df
```

```python
import json
import uuid
from pathlib import Path

src_path = Path("/home/zhangxin/Documents/imc_dashboard/logfiles/2026-04-07_22-03-51.log")
out_path = Path("/home/zhangxin/Documents/imc_dashboard/logfiles/converted_2026-04-07_22-03-51.log")

lines = src_path.read_text(encoding="utf-8").splitlines()

# 1) 找三段的起始行号
i_sandbox = next(i for i, s in enumerate(lines) if s.strip() == "Sandbox logs:")
i_acts = next(i for i, s in enumerate(lines) if s.strip() == "Activities log:")
i_trades = next(i for i, s in enumerate(lines) if s.strip() == "Trade History:")

# 2) 解析 Sandbox logs: 多个 JSON 对象块 -> logs(list[dict])
logs = []
buf = []
in_obj = False
for s in lines[i_sandbox + 1 : i_acts]:
    st = s.strip()
    if not st:
        continue
    if st == "{":
        in_obj = True
        buf = ["{"]
        continue
    if in_obj:
        buf.append(s)
        if st == "}":
            try:
                logs.append(json.loads("\n".join(buf)))
            except json.JSONDecodeError as e:
                raise ValueError(f"解析 sandbox 单条 JSON 失败: {e}\n块内容:\n{''.join(buf[:50])}")
            in_obj = False
            buf = []
# 若文件结尾异常导致没收尾，也可以在这里处理 buf

# 3) 解析 Activities log: 原样拼成一个 CSV 字符串 -> activitiesLog(str)
# activitiesLog 需要包含表头那行，所以从 "Activities log:" 下一行开始，到 "Trade History:" 前为止
activitiesLog = "\n".join([s.rstrip("\n") for s in lines[i_acts + 1 : i_trades] if s.strip()])

# 4) 解析 Trade History: JSON 数组 -> tradeHistory(list[dict])
trade_text = "\n".join([s for s in lines[i_trades + 1 :] if s.strip()])
# trade_text 里应以 '[' 开头
j0 = trade_text.find("[")
if j0 == -1:
    tradeHistory = []
else:
    raw = trade_text[j0:]
    # 去掉 } 或 ] 前的多余逗号：...,} -> ...} 以及 ...,] -> ...]
    cleaned = re.sub(r",\s*([}\]])", r"\1", raw)
    tradeHistory = json.loads(cleaned)

# 5) 组装成 58413.log 那种单行 JSON
out_obj = {
    "submissionId": str(uuid.uuid4()),
    "activitiesLog": activitiesLog,
    "logs": logs,
    "tradeHistory": tradeHistory,
}

out_path.write_text(json.dumps(out_obj, ensure_ascii=False), encoding="utf-8")

print("written:", out_path)
print("logs:", len(logs), "tradeHistory:", len(tradeHistory), "activitiesLog_lines:", activitiesLog.count("\n") + 1)
```
